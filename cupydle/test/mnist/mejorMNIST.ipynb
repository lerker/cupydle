```python
ax = plt.subplot(111)
ax.set_xscale("log")
ax.set_yscale("log")
x = np.asarray(list(range(0, len(validacion))))
y = validacion
plt.errorbar(x, y, yerr=5)
ax.set_ylim(ymin=0.1)
ax.set_title('Errorbars go negative')
plt.show()
```

```python
line,caps,bars=plt.errorbar(
    list(range(0,len(validacion))),     # X
    validacion,    # Y
    yerr=10,        # Y-errors
    fmt="rs--",    # format line like for plot()
    linewidth=0.3,   # width of plot line
    elinewidth=0.05,# width of error bar line
    ecolor='k',    # color of error bar
    capsize=0.05,     # cap length for error bar
    capthick=0.005   # cap thickness for error bar
    )

plt.setp(line,label="My error bars")#give label to returned line
plt.legend(numpoints=1,             #Set the number of markers in label
           loc=('upper left'))      #Set label location
```
